```python
import pandas as pd
df = pd.read_csv('D:/CDW/CDW_Internship/all_kindle_review.csv')
reviews = df[['reviewText', 'rating']]
```

```python
reviews
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(reviews)):
    review=re.sub('[^a-zA-z]',' ',reviews['reviewText'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## Output Features
y=pd.get_dummies(reviews['rating'])
```

```python
y=y.iloc[:,0].values
```

```python
y
```

```python
#Create Bag of Words
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_train),len(y_train)
```

```python
## independent features
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
X_train
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
sentimental_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=sentimental_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
len(corpus)
```

```python
len(X_train),len(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train=tv.fit_transform(X_train).toarray()
X_test=tv.transform(X_test).toarray()
```

```python
X_train
```

```python
tv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model = MultinomialNB().fit(X_train, y_train)
```

```python
#prediction
y_pred=spam_tfidf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```
