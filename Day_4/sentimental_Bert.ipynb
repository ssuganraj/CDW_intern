```python

from transformers import BertTokenizer, BertModel
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
# Load pre-trained BERT model and tokenizer
model_name = 'bert-base-uncased'  # You can choose other BERT models
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python

# Load a Sentence Transformer model for cosine similarity
sentence_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
def analyze_sentiment(sentence):
    """
    Analyzes the sentiment of a given sentence using BERT and cosine similarity.

    Args:
      sentence: The input sentence.

    Returns:
      "Positive", "Negative", or "Neutral" based on the sentiment analysis.
    """

    # Encode the input sentence and reference sentences
    positive_reference = "This is fantastic!"
    negative_reference = "This is terrible!"

    sentence_embedding = sentence_model.encode(sentence)
    positive_embedding = sentence_model.encode(positive_reference)
    negative_embedding = sentence_model.encode(negative_reference)
    # Calculate cosine similarity
    positive_similarity = util.cos_sim(sentence_embedding, positive_embedding)
    negative_similarity = util.cos_sim(sentence_embedding, negative_embedding)

    # Determine sentiment based on similarity scores
    if positive_similarity > negative_similarity:
        return "Positive"
    elif negative_similarity > positive_similarity:
        return "Negative"
    else:
        return "Neutral"  # Or handle as needed
```

```python
# Example usage
input_sentence = "I hate this product!"
sentiment = analyze_sentiment(input_sentence)
print(f"The sentiment of '{input_sentence}' is: {sentiment}")

input_sentence = "This is the love movie ever!"
sentiment = analyze_sentiment(input_sentence)
print(f"The sentiment of '{input_sentence}' is: {sentiment}")

input_sentence = "The weather is good today"
sentiment = analyze_sentiment(input_sentence)
print(f"The sentiment of '{input_sentence}' is: {sentiment}")
```
